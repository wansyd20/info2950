# Cleaning and Loading Amazon Review Datasets
We are using the Electronics category to demonstrate how we loaded and narrowed down the review dataset process. We performed the same processes to three other Amazon review categories (Beauty, Office Products, Apparel) in separate JupyterNotebook files. We loaded those here and merged all the cleaned datasets into one comprehensive dataset, which we have provided in CMS. 

```python
electronics = pd.read_csv('/Users/michellehui/info2950_project/amazon_reviews_us_Electronics_v1_00.tsv', sep = '\t', error_bad_lines=False)
```

```python
# keeping reviews written in 2015
electronics['review_date'] = pd.to_datetime(electronics['review_date'])
electronics = electronics.loc[electronics['review_date'].dt.year == 2015]
```

```python
# dropping marketplace column
# electronics = electronics.drop(columns= ['marketplace'], axis=1)
electronics.head(-5)
```

```python
# randomly narrowing down 800,000 observations to 10,000
electronics = electronics.sample(n=10000)
electronics.reset_index()
```

```python
# downloading cleaned csv file
electronics.to_csv('amazonreviews_electronics.csv',index=False)
```
