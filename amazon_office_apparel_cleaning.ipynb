```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv("amazon_reviews_office_products.tsv", sep = "\t", error_bad_lines= False)
```

```python
office_products = raw_data.copy()
```

```python
office_products.head()
```

```python
office_products['review_date'] = pd.to_datetime(office_products['review_date'])
```

```python
office_products.loc[:, 'review_date'] = office_products[office_products['review_date'].dt.year == 2015]
```

```python
office_products = office_products.dropna(subset=['review_date'])
```

```python
office_products.head()
```

```python
office_products.to_csv('amazon_office_products_clean1.csv', index=False)
```

```python
office_products_sample = office_products.sample(n=10000).reset_index()
office_products_sample = office_products_sample.drop(columns = ['index', 'marketplace'])
```

```python
office_products_sample
```

```python
office_products_sample.to_csv('amazon_reviews_office_products.csv', index = False)
```
